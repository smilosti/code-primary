```python
import pandas as pd

def clean_data(df):
    # Rename column 'Control Identifier' to 'ctrl_id'
    df = df.rename(columns={'Control Identifier': 'ctrl_id'})
    # Rename column 'Control (or Control Enhancement) Name' to 'ctrl_title'
    df = df.rename(columns={'Control (or Control Enhancement) Name': 'ctrl_title'})
    # Rename column 'Control Text' to 'ctrl_specification'
    df = df.rename(columns={'Control Text': 'ctrl_specification'})
    # Rename column 'Discussion' to 'ctrl_guidance'
    df = df.rename(columns={'Discussion': 'ctrl_guidance'})
    # Rename column 'ctrl_guidance' to 'ctrl_discussion'
    df = df.rename(columns={'ctrl_guidance': 'ctrl_discussion'})
    # Rename column 'Related Controls' to 'ctrl_related_ctrls'
    df = df.rename(columns={'Related Controls': 'ctrl_related_ctrls'})
    return df

# Loaded variable 'df' from URI: /workspaces/code/services/typedb/sp800-53r5-control-catalog.xlsx
df = pd.read_excel(r'/workspaces/code/services/typedb/sp800-53r5-control-catalog.xlsx')

df_clean = clean_data(df.copy())
df_clean.head()
```
